```python

# Sort the DataFrame based on the "population_value" column in descending order
sorted_data = df.sort_values(by='World Population Percentage', ascending=False)

# Print the sorted DataFrame
sorted_data.head(10)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True)

plt.rcParams['figure.figsize'] = (30,9)

plt.show()
```

```python
df.groupby('Continent').mean().sort_values(by="World Population Percentage",ascending=False)
```
